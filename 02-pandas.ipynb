## apply

```python
df['Name_len']= df['Name'].apply(lambda x : len(x))
df[['Name','Name_len']].head(3)
```

```python
df.drop('Name_len', axis=1, inplace=True)
df.head()
```

```python
lambda_len = lambda x : len(x)

df['Name_len']= df['Name'].apply(lambda_len)
df.head()
```

```python
df['Child_Adult'] = df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult' )
df[['Age','Child_Adult']].head(8)
```

```python
lambda_age = lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 'Elderly')

df['Age_cat'] = df['Age'].apply(lambda_age)
df['Age_cat'].value_counts()
```

```python
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

df['Age_cat'] = df['Age'].apply(lambda x : get_category(x))
df[['Age','Age_cat']].head()
```
