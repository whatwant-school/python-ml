## K-Fold

```python
iris_df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split

features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
label = 'label'
n_splits = 5

K_clf = DecisionTreeClassifier(random_state=156)
kfold = KFold(n_splits=n_splits)

X_train, X_test, y_train, y_test = train_test_split(iris_df[features], iris_df[label], test_size=0.2, random_state=11)
X_train.info()
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

n_iter = 0
cv_accuracy = []
for train_index, test_index  in kfold.split(X_train):
    K_X_train, K_X_test = X_train.iloc[train_index], X_train.iloc[test_index]
    K_y_train, K_y_test = y_train.iloc[train_index], y_train.iloc[test_index]

    K_clf.fit(K_X_train , K_y_train)    
    pred = K_clf.predict(K_X_test)
    
    n_iter += 1

    accuracy = np.round(accuracy_score(K_y_test, pred), 4)
    
    train_size = K_X_train.shape[0]
    test_size = K_X_test.shape[0]
    
    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
          .format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))
    
    cv_accuracy.append(accuracy)
```

## Stratified K-Fold

```python
iris_df['label'].value_counts()
```

```python
n_splits = 3

kfold = KFold(n_splits=n_splits)

n_iter =0
for train_index, test_index  in kfold.split(iris_df):
    n_iter += 1
    
    y_train, y_test = iris_df['label'].iloc[train_index], iris_df['label'].iloc[test_index]

    print('## 교차 검증: {0}'.format(n_iter))
    print('학습 레이블 데이터 분포:\n', y_train.value_counts())
    print('검증 레이블 데이터 분포:\n', y_test.value_counts())
    print()
```

```python
from sklearn.model_selection import StratifiedKFold

n_splits = 3
skf = StratifiedKFold(n_splits=n_splits)

n_iter=0
for train_index, test_index in skf.split(iris_df, iris_df['label']):
    n_iter += 1
    
    label_train= iris_df['label'].iloc[train_index]
    label_test= iris_df['label'].iloc[test_index]
    
    print('## 교차 검증: {0}'.format(n_iter))
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
```

## cross_val_score()

```python
from sklearn.model_selection import cross_val_score

features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
label = 'label'

clf = DecisionTreeClassifier(random_state=156)

scores = cross_val_score(clf, iris_df[features], iris_df[label], scoring='accuracy', cv=3)

print('교차 검증별 정확도:',np.round(scores, 4))
print('평균 검증 정확도:', np.round(np.mean(scores), 4))
```

## GridSarchCV

```python
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
label = 'label'

X_train, X_test, y_train, y_test = train_test_split(iris_df[features], iris_df[label], test_size=0.2, random_state=11)

clf = DecisionTreeClassifier()

from sklearn.model_selection import GridSearchCV

parameters = {'max_depth':[1,2,3], 'min_samples_split':[2,3]}

grid_cv = GridSearchCV(clf, param_grid=parameters, cv=3, refit=True)
grid_cv.fit(X_train, y_train)

scores_df = pd.DataFrame(grid_cv.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score',
           'split0_test_score', 'split1_test_score', 'split2_test_score']]

print('GridSearchCV 최적 파라미터:', grid_cv.best_params_)
print('GridSearchCV 최고 정확도: {0:.4f}'.format(grid_cv.best_score_))
```

```python
estimator = grid_cv.best_estimator_

pred = estimator.predict(X_test)
print('테스트 데이터 세트 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```
